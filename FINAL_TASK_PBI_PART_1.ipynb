```python
# Korelasi numerik (point-biserial) & Mutual Information kategorikal
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from scipy.stats import pointbiserialr

df = app_train.copy()
y = df["TARGET"].astype(int)

# Numerik
num_cols = df.select_dtypes(include=[np.number]).columns.drop("TARGET")
corr_rows = []
for c in num_cols:
    x = df[c]
    x = x.replace([np.inf, -np.inf], np.nan)
    if x.notna().sum() > 100:  # minimal non-null
        r = pointbiserialr(y.loc[x.notna()], x.loc[x.notna()])[0]
        corr_rows.append((c, r))
corr_df = (pd.DataFrame(corr_rows, columns=["feature","point_biserial_corr"])
           .dropna().sort_values("point_biserial_corr", key=lambda s: s.abs(), ascending=False))
print("Top 15 numerik vs TARGET:")
print(corr_df.head(15))

# Kategorikal
cat_cols = df.select_dtypes(include=["object","category"]).columns
X_cat = pd.get_dummies(df[cat_cols], dummy_na=True, drop_first=False)
mi = mutual_info_classif(X_cat.fillna(0), y, discrete_features=True, random_state=42)
mi_df = pd.DataFrame({"feature": X_cat.columns, "mutual_info": mi}).sort_values("mutual_info", ascending=False)
print("\nTop 15 kategorikal (MI) vs TARGET:")
print(mi_df.head(15))
```

```python
# ============================
# Strategi Missing Value Lengkap & Konsisten - PART 1
# ============================

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

print("🔧 Memulai perbaikan Missing Value dengan strategi konsisten...")

# 1. Kolom apartemen untuk indicator (yang memang bermakna jika missing = 0)
apartment_cols = [
    'COMMONAREA_AVG', 'LIVINGAPARTMENTS_AVG', 'NONLIVINGAPARTMENTS_AVG',
    'COMMONAREA_MODE', 'LIVINGAPARTMENTS_MODE', 'NONLIVINGAPARTMENTS_MODE',
    'COMMONAREA_MEDI', 'LIVINGAPARTMENTS_MEDI', 'NONLIVINGAPARTMENTS_MEDI'
]

# 2. Tambahkan missing indicator untuk kolom apartemen SEBELUM imputasi
for col in apartment_cols:
    if col in app_train.columns:
        app_train[col + "_MISSING"] = app_train[col].isnull().astype(int)

# 3. Identifikasi kolom
id_cols = [col for col in app_train.columns if col.startswith('SK_ID')]
target_col = ['TARGET'] if 'TARGET' in app_train.columns else []
all_num_cols = app_train.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [col for col in all_num_cols if col not in id_cols + target_col + [c + "_MISSING" for c in apartment_cols]]
cat_cols = app_train.select_dtypes(include=['object']).columns.tolist()

# 4. Pisahkan kolom numerik apartemen dari numerik lainnya
apartment_num_cols = [col for col in apartment_cols if col in num_cols]
other_num_cols = [col for col in num_cols if col not in apartment_num_cols]

print(f"🔹 Kolom apartemen: {len(apartment_num_cols)}")
print(f"🔹 Kolom numerik lainnya: {len(other_num_cols)}")
print(f"🔹 Kolom kategorikal: {len(cat_cols)}")

# 5. Definisikan imputers
preprocessor = ColumnTransformer(
    transformers=[
        ('num_apartment', SimpleImputer(strategy='constant', fill_value=0), apartment_num_cols),
        ('num_other', SimpleImputer(strategy='median'), other_num_cols),
        ('cat', SimpleImputer(strategy='constant', fill_value='Unknown'), cat_cols)
    ],
    remainder='passthrough'  # Kolom lain (ID, target, indicator) tidak diimputasi
)

# 6. Fit dan transform
print("⏳ Melakukan imputasi...")
app_train_imputed = preprocessor.fit_transform(app_train)

# 7. Dapatkan nama kolom setelah transformasi
feature_names = (apartment_num_cols + other_num_cols + cat_cols +
                [col for col in app_train.columns if col not in apartment_num_cols + other_num_cols + cat_cols])

# 8. Konversi kembali ke DataFrame
app_train_clean = pd.DataFrame(app_train_imputed, columns=feature_names)

# 9. Pastikan tipe data konsisten
for col in apartment_num_cols + other_num_cols:
    app_train_clean[col] = app_train_clean[col].astype(float)

for col in cat_cols:
    app_train_clean[col] = app_train_clean[col].astype(str)

if 'TARGET' in app_train_clean.columns:
    app_train_clean['TARGET'] = app_train_clean['TARGET'].astype(int)

# 10. DROP kolom yang tidak perlu (PHASE 1)
cols_to_drop_phase1 = [
    # Document Flags
    'FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
    'FLAG_DOCUMENT_6', 'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9',
    'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13',
    'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17',
    'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21',

    # Credit Bureau Requests
    'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY',
    'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_MON',
    'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR',

    # Mostly constant flags
    'FLAG_MOBIL', 'FLAG_CONT_MOBILE'
]

app_train_clean = app_train_clean.drop(columns=[col for col in cols_to_drop_phase1 if col in app_train_clean.columns], errors='ignore')

# 11. Verifikasi final
print(f"\n✅ Transformasi selesai. Shape final: {app_train_clean.shape}")
print(f"🔹 Total missing values: {app_train_clean.isnull().sum().sum()}")
print(f"🔹 Duplicates: {app_train_clean.duplicated().sum()}")

# 12. Simpan
cleaned_data_path = "cleaned_app_train2.csv"
app_train_clean.to_csv(cleaned_data_path, index=False)
print(f"💾 Disimpan sebagai: {cleaned_data_path}")
```
